## Adat betöltése

```python
import pandas as pd
```

```python
percentify = lambda x: (x * 100).round(1).astype(str) + "%"
```

```python
posts = pd.read_csv("data.csv")
```

## Adatbázis mérete

```python
len(posts)
```

```python
posts.iloc[0:3]
```

```python
RESZVETELI_OSZLOPOK = [
    "share_count",
    "reaction_count",
    "comment_count"
]
```

```python
RATINGS = ["mostly false", "mixture of true and false", "mostly true", "no factual content"]
FACTUAL_RATINGS = ["mostly false", "mixture of true and false", "mostly true"]
```

```python
category_grp = posts.groupby("Category")
page_grp = posts.groupby([ "Category", "Page" ])
type_grp = posts.groupby([ "Category", "Page", "Post Type" ])
```

## Kategória szerinti értékelés

```python
ratings = category_grp["Rating"]
rating_by_category = ratings.value_counts().unstack()[RATINGS].fillna(0)
rating_by_category["total"] = rating_by_category.sum(axis=1)
rating_by_category = rating_by_category.rename_axis("Kategória")
rating_by_category = rating_by_category.rename_axis("Értékelés", axis="columns")
rating_by_category
```

**Százalékosan:**

```python
rating_by_category_perc = (rating_by_category[RATINGS].T / rating_by_category[RATINGS].sum(axis=1)).T\
    .pipe(percentify)
rating_by_category_perc = rating_by_category_perc.rename_axis("Kategória")
rating_by_category_perc = rating_by_category_perc.rename_axis("Értékelés", axis="columns")
rating_by_category_perc
```

**Százalékosan, a politikai hírek:**

```python
factual_perc = (rating_by_category[FACTUAL_RATINGS].T / rating_by_category[FACTUAL_RATINGS].sum(axis=1)).T\
    .pipe(percentify)
factual_perc = factual_perc.rename_axis("Kategória")
factual_perc = factual_perc.rename_axis("Értékelés", axis="columns")
factual_perc
```

## Értékelés oldalak szerint

```python
rating_by_page = page_grp["Rating"].value_counts().unstack()[RATINGS].fillna(0)
rating_by_page["total"] = rating_by_page.sum(axis=1)
rating_by_page = rating_by_page.rename_axis(["Kategória", "Oldalak"])
rating_by_page = rating_by_page.rename_axis("Értékelés", axis="columns")
rating_by_page
```

**Százalékosan:**

```python
rating_by_page_perc = (rating_by_page[RATINGS].T / rating_by_page[RATINGS].sum(axis=1)).T\
    .pipe(percentify)
rating_by_page_perc = rating_by_page_perc.rename_axis(["Kategória", "Oldalak"])
rating_by_page_perc = rating_by_page_perc.rename_axis("Értékelés", axis="columns")
rating_by_page_perc
```

**Százalékosan, a politikai hírek:**

```python
page_perc = (rating_by_page[FACTUAL_RATINGS].T / rating_by_page[FACTUAL_RATINGS].sum(axis=1)).T\
    .pipe(percentify)
page_perc = page_perc.rename_axis(["Kategória", "Oldalak"])
page_perc = page_perc.rename_axis("Értékelés", axis="columns")
page_perc
```

## Hírek száma dátum szerint

```python
posts_by_date_by_category = category_grp["Date Published"].value_counts().unstack()
posts_by_date_by_category["Avg. Per Day"] = posts_by_date_by_category.mean(axis=1).round(0)
posts_by_date_by_category = posts_by_date_by_category.rename_axis("Kategória")
posts_by_date_by_category = posts_by_date_by_category.rename_axis("Megjelenés dátuma", axis="columns")
posts_by_date_by_category
```

```python
posts_by_date_by_page = page_grp["Date Published"].value_counts().unstack()
posts_by_date_by_page["Avg. Per Day"] = posts_by_date_by_page.mean(axis=1).round(0)
posts_by_date_by_page = posts_by_date_by_page.rename_axis(["Kategória", "Oldalak"])
posts_by_date_by_page = posts_by_date_by_page.rename_axis("Megjelenés dátuma", axis="columns")
posts_by_date_by_page
```

## Értékelés típus alapján

```python
rating_by_post_type = type_grp["Rating"].value_counts().unstack()[RATINGS].fillna(0)
rating_by_post_type["total"] = rating_by_post_type.sum(axis=1)
rating_by_post_type = rating_by_post_type.rename_axis(["Kategória", "Oldalak", "Típus"])
rating_by_post_type = rating_by_post_type.rename_axis("Értékelés", axis="columns")
rating_by_post_type
```

## Megosztások a kapcsolódó vs. nem kapcsolódó hírek szerint

```python
grp = posts.groupby([ "Category", "Page", posts["Rating"] == "no factual content" ])
out = pd.DataFrame({
    "medián": grp["share_count"].median(),
    "átlag": grp["share_count"].mean()
}).round()\
    .unstack().stack(level=0).rename(columns={True: "nem kapcsolódó", False: "kapcsolódó"})
out = out.rename_axis(["Kategória", "Oldalak", ""])
out = out.rename_axis("Értékelés", axis="columns")
out
```

## Megosztások a többnyire igaz vs összes többi a jellemzően álhír oldalakra

```python
grp = posts.groupby([ "Category", "Page", posts["Rating"] == "mostly true" ])
out = pd.DataFrame({
    "medián": grp["share_count"].median(),
    "átlag": grp["share_count"].mean()
}).round()\
    .unstack().stack(level=0).rename(columns={True: "mostly true", False: "többi"})\
    [[ "mostly true", "többi" ]].loc[["left", "right"]]
out = out.rename_axis(["Kategória", "Oldalak", ""])
out = out.rename_axis("Értékelés", axis="columns")
out
```
